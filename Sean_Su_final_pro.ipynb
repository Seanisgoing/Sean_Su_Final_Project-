```python
import pandas as pd
import numpy as np
import sqlite3
import plotly.express as px
```

```python
df = pd.read_csv("https://github.com/Seanisgoing/Sean_Su_Final_Project-/raw/main/tv_shows.csv")
```

```python
num_rols = df.shape[0]
num_cols = df.shape[1]
print(num_rols, num_cols)
```

```python
df.info()
```

```python
dfclean = df[(df["Age"].notna()) & (df["IMDb"].notna())]
```

```python
df.head(50)
```

```python
top20_shows_imdb = dfclean.sort_values(by= "IMDb", ascending= False)
top20_shows_imdb.head(20)
```

```python
top20_shows_rt = dfclean.sort_values(by= "Rotten Tomatoes", ascending= False)
top20_shows_rt.head(20)
```

```python
df_by_age = dfclean.groupby(by= ["Age"], as_index= False).count()
df_by_age = df_by_age.rename(columns= {"Unnamed: 0": "Number of shows by age"})
df_by_age
```

```python
df_nfshows = len(dfclean[dfclean["Netflix"] == 1])
df_hlshows = len(dfclean[dfclean["Hulu"] == 1])
df_prime = len(dfclean[dfclean["Prime Video"] == 1])
df_disney = len(dfclean[dfclean["Disney+"]==1])
print(f"Netflix shows: {df_nfshows}")
print(f"Hulu shows: {df_hlshows}")
print(f"Prime shows: {df_prime}")
print(f"Disney+ shows: {df_disney}")
```
